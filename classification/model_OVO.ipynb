```python
raw.notch_filter(freqs=50)
raw.filter(.5, 30., fir_design='firwin')

picks = mne.pick_types(raw.info, meg=False, eeg=True, eog=False, stim=False,
                       exclude='bads')  
tmin, tmax = 0, 5
event_id = dict({'Fist': 1, 'Hook': 2, 'Index': 3, 'Open': 4, 'Thumb': 5})

epochs = mne.Epochs(raw, events, event_id, tmin, tmax, proj=True, picks=picks,
                baseline=None, preload=True)
```

```python
labels = epochs.events[:,-1]
data = epochs.get_data()
print(data.shape, labels.shape)
```

```python
import pywt

# signal is decomposed to level 5 with 'db4' wavelet

def wpd(X): 
    coeffs = pywt.WaveletPacket(X,'db4',mode='symmetric',maxlevel=5)
    return coeffs
             
def feature_bands(x):
    C = wpd(x[0, 0, :])
    coeffs_sample = [node.data for node in C.get_level(5, 'natural')]
    jumlah_koefisien = len(coeffs_sample[0])  
    
    Bands = np.empty((8,x.shape[0],x.shape[1],jumlah_koefisien)) 
    
    for i in range(x.shape[0]):
        for ii in range(x.shape[1]):
             pos = []
             C = wpd(x[i,ii,:]) 
             pos = np.append(pos,[node.path for node in C.get_level(5, 'natural')])
             for b in range(1,9):
                 Bands[b-1,i,ii,:] = C[pos[b]].data
        
    return Bands

wpd_data = feature_bands(data)
print(wpd_data.shape)
```

```python
# Library
import numpy as np
import pandas as pd
from sklearn import preprocessing
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, cohen_kappa_score, precision_score, recall_score
from sklearn.model_selection import StratifiedKFold, GridSearchCV, train_test_split
from mne.decoding import CSP
from sklearn.multiclass import OneVsOneClassifier
from sklearn.calibration import CalibratedClassifierCV

# Data preparation
data_comb = [wpd_data[i] for i in range(wpd_data.shape[0])]
labels_copy = labels.copy()

indices = np.arange(data_comb[0].shape[0])

# Membagi data menjadi training dan testing
indices_train, indices_test = train_test_split(
    indices, test_size=0.2, stratify=labels_copy, random_state=42
)
data_comb_train = [data_comb[i][indices_train] for i in range(len(data_comb))]
data_comb_test = [data_comb[i][indices_test] for i in range(len(data_comb))]

wpd_data_train = np.asarray(data_comb_train)
wpd_data_test = np.asarray(data_comb_test)
labels_train = labels_copy[indices_train]
labels_test = labels_copy[indices_test]

# Inisialisasi dan fitting CSP pada data training
Csp = [
    CSP(n_components=4, reg=None, log=True, norm_trace=False)
    for _ in range(wpd_data_train.shape[0])
]
for x in range(wpd_data_train.shape[0]):
    Csp[x].fit(wpd_data_train[x, ...], labels_train)

# Menggabungkan fitur CSP untuk data training
X_full_train = np.concatenate(
    [Csp[x].transform(wpd_data_train[x, :, :, :]) for x in range(wpd_data_train.shape[0])],
    axis=1,
)

# Standarisasi data
ss = preprocessing.StandardScaler()
X_full_train = ss.fit_transform(X_full_train)

# Inisialisasi list untuk menyimpan model dan metrik evaluasi
svms = []
acc = []
ka = []
prec = []
recall = []

# Parameter grid untuk GridSearchCV
param_grid = {
    'base_estimator__estimator__C': [0.1, 1, 1.5, 0.9],
    'base_estimator__estimator__gamma': [1, 0.1, 0.01, 0.001],
}

# Inisialisasi StratifiedKFold
cv = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)

# Loop cross-validation
for train_idx, val_idx in cv.split(X_full_train, labels_train):
    X_train, X_val = X_full_train[train_idx], X_full_train[val_idx]
    y_train, y_val = labels_train[train_idx], labels_train[val_idx]

    # Inisialisasi model SVM
    svm = SVC(kernel='rbf')

    # Menggunakan OneVsOneClassifier untuk OVO
    ovo_clf = OneVsOneClassifier(svm)

    # Kalibrasi model untuk mendapatkan probabilitas
    calibrated_clf = CalibratedClassifierCV(ovo_clf, cv=5, method='sigmoid')

    # Inisialisasi GridSearchCV dengan model terkalibrasi
    grid_search = GridSearchCV(
        calibrated_clf, param_grid, refit=True, verbose=0, cv=5
    )

    # Fit GridSearchCV pada data training
    grid_search.fit(X_train, y_train)

    best_svm = grid_search.best_estimator_

    svms.append(best_svm)

    # Predict pada data validation
    y_pred_val = best_svm.predict(X_val)

    # Hitung dan simpan metrik evaluasi pada data validation
    acc.append(accuracy_score(y_val, y_pred_val))
    ka.append(cohen_kappa_score(y_val, y_pred_val))
    prec.append(precision_score(y_val, y_pred_val, average='weighted'))
    recall.append(recall_score(y_val, y_pred_val, average='weighted'))

# Evaluasi pada data testing dengan model terbaik
best_idx = np.argmax(acc)
final_model = svms[best_idx]

# Ekstraksi fitur CSP dan standarisasi untuk data testing
X_test = np.concatenate(
    [Csp[x].transform(wpd_data_test[x, :, :, :]) for x in range(wpd_data_test.shape[0])],
    axis=1,
)
X_test = ss.transform(X_test)

# Predict pada data testing
y_pred_test = final_model.predict(X_test)

# Metrik evaluasi pada data testing
test_acc = accuracy_score(labels_test, y_pred_test)
test_ka = cohen_kappa_score(labels_test, y_pred_test)
test_prec = precision_score(labels_test, y_pred_test, average='weighted')
test_recall = recall_score(labels_test, y_pred_test, average='weighted')

# Simpan metrik evaluasi pada data testing ke dalam dictionary
test_scores = {
    'Accuracy': test_acc,
    'Kappa': test_ka,
    'Precision': test_prec,
    'Recall': test_recall,
}

scores = {'Accuracy': acc, 'Kappa': ka, 'Precision': prec, 'Recall': recall}

Es = pd.DataFrame(scores)
avg = {
    'Accuracy': [np.mean(acc)],
    'Kappa': [np.mean(ka)],
    'Precision': [np.mean(prec)],
    'Recall': [np.mean(recall)],
}
Avg = pd.DataFrame(avg)

T = pd.concat([Es, Avg], ignore_index=True)

T.index = [
    'F1','F2','F3','F4','F5','F6','F7','F8','F9','F10','Avg',]
T.index.rename('Fold', inplace=True)

print("Metrik Evaluasi per Fold (Validation Set):")
print(T)

print("\nMetrik Evaluasi pada Data Testing:")
print("Testing Accuracy:", test_scores['Accuracy'])
print("Testing Cohen's Kappa:", test_scores['Kappa'])
print("Testing Precision:", test_scores['Precision'])
print("Testing Recall:", test_scores['Recall'])
```

```python
# Import library yang diperlukan
import numpy as np
from sklearn import preprocessing
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, cohen_kappa_score, precision_score, recall_score
from sklearn.model_selection import StratifiedKFold, GridSearchCV, train_test_split
from mne.decoding import CSP
from sklearn.multiclass import OneVsRestClassifier

# Persiapan data (asumsi wpd_data dan labels sudah didefinisikan sebelumnya)
labels_copy = labels.copy()

indices = np.arange(wpd_data.shape[1])

# Membagi data menjadi training dan testing
indices_train, indices_test = train_test_split(indices, test_size=0.2, stratify=labels_copy, random_state=42)
wpd_data_train = wpd_data[:, indices_train, :, :]
wpd_data_test = wpd_data[:, indices_test, :, :]
labels_train = labels_copy[indices_train]
labels_test = labels_copy[indices_test]

# Inisialisasi dan fitting CSP pada data training
Csp = [CSP(n_components=4, reg=None, log=True, norm_trace=False) for _ in range(wpd_data.shape[0])]
for x in range(wpd_data.shape[0]):
    Csp[x].fit(wpd_data_train[x, :, :, :], labels_train)

# Menggabungkan fitur CSP untuk data training
X_full_train = np.concatenate([Csp[x].transform(wpd_data_train[x, :, :, :]) for x in range(wpd_data.shape[0])], axis=1)

# Inisialisasi StandardScaler dan fit pada data training
ss = preprocessing.StandardScaler()
X_full_train = ss.fit_transform(X_full_train)

# Inisialisasi list untuk menyimpan metrik evaluasi
acc = []
ka = []
prec = []
recall = []

# Parameter grid untuk GridSearchCV
param_grid = {
    'estimator__C': [0.1, 1, 1.5, 0.9], 
    'estimator__gamma': [1, 0.1, 0.01, 0.001]
}

# Inisialisasi StratifiedKFold
cv = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)

# Loop cross-validation
for train_idx, val_idx in cv.split(X_full_train, labels_train):
    
    X_train, X_val = X_full_train[train_idx], X_full_train[val_idx]
    y_train, y_val = labels_train[train_idx], labels_train[val_idx]
    
    # Inisialisasi model SVM dengan OVA
    svm = SVC(kernel='rbf', probability=True)
    ova_clf = OneVsRestClassifier(svm)
    
    # Inisialisasi GridSearchCV dengan OVA
    grid_search = GridSearchCV(ova_clf, param_grid, refit=True, verbose=0, cv=5)
    
    # Fit GridSearchCV pada data training
    grid_search.fit(X_train, y_train)
    
    # Predict pada data validation
    y_pred_val = grid_search.predict(X_val)
    
    # Hitung dan simpan metrik evaluasi pada data validation
    acc.append(accuracy_score(y_val, y_pred_val))
    ka.append(cohen_kappa_score(y_val, y_pred_val))
    prec.append(precision_score(y_val, y_pred_val, average='weighted'))
    recall.append(recall_score(y_val, y_pred_val, average='weighted'))

# Setelah cross-validation, fit model pada seluruh data training
final_svm = SVC(kernel='rbf', probability=True, C=grid_search.best_params_['estimator__C'], gamma=grid_search.best_params_['estimator__gamma'])
final_ova_clf = OneVsRestClassifier(final_svm)
final_ova_clf.fit(X_full_train, labels_train)

# Persiapan data testing
X_test = np.concatenate([Csp[x].transform(wpd_data_test[x, :, :, :]) for x in range(wpd_data.shape[0])], axis=1)
X_test = ss.transform(X_test)

# Predict pada data testing
y_pred_test = final_ova_clf.predict(X_test)

# Metrik evaluasi pada data testing
test_acc = accuracy_score(labels_test, y_pred_test)
test_ka = cohen_kappa_score(labels_test, y_pred_test)
test_prec = precision_score(labels_test, y_pred_test, average='weighted')
test_recall = recall_score(labels_test, y_pred_test, average='weighted')

print("Pendekatan OVA:")
print("Testing Accuracy:", test_acc)
print("Testing Cohen's Kappa:", test_ka)
print("Testing Precision:", test_prec)
print("Testing Recall:", test_recall)
```
